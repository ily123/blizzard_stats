```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_fp = 'data/bfa_patch83_dungeon_mob_count.csv'
df = pd.read_csv(data_fp)
df
```

```python
df['ratio'] = df.required_count_teeming/df.required_count_normal
```

```python
df = df.sort_values('ratio', ascending=False)
```

```python
plt.barh(list(range(0,len(df))), df.ratio)
plt.xlim(1,1.21)
plt.title('Most dungeons require 20% extra mob count on teeming\nKR and UD are the only exceptions')
plt.xticks([1, 1.1, 1.2], ['0%', '10%', '20%'])
plt.yticks(list(range(0,len(df))), df.name)

plt.xlabel('Extra % mob count needed on teeming (compared to normal)')
```

```python
df['ratio2'] = df.required_count_normal/df.total_available_count_normal
df = df.sort_values('ratio2', ascending=False)
plt.barh(list(range(0,len(df))), df.ratio2)
plt.yticks(list(range(0,len(df))), df.name);
```
